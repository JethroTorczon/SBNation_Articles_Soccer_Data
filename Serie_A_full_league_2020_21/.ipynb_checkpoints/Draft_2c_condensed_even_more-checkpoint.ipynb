## Part 2 - PLOTTING the data

### Before actually creating the plots themselves, it is helpful to first define functions that can be used for customizing the formatting of any graph type that will be used more than once

```python
def label_scatter_point(x,
                        y,
                        val,   #  point labels' source column
                        ax     #  plt.gca()
                       ):
    
    a = pd.concat({'x': x,
                   'y': y,
                   'val': val},
                  
                  axis = 1    # I think this clarifies that these are COLUMNS, not rows)
                 )

    
    for i, point in a.iterrows():

        ax.text(point['x'] + .04,
                point['y'] + .15,  # PLUS to shift UP
                point['val']
               )
```

#### example input to run the "label_scatter_point" function:

label_scatter_point( 

                    SerieA_table_df.GF,
    
                    SerieA_table_df.GA,
                    
                    SerieA_table_df.Squad,   # "Squad" column = point label source
                    
                    plt.gca()
                    
                   )

## Scatterplot - Goals FOR vs Goals AGAINST (all Serie A teams)

```python
ax = sns.lmplot('GF',                       # horizontal axis
                'GA',                       # vertical axis
                
                data = SerieA_table_df,     # data source
                
                #fit_reg = False,            # do NOT fix a regression line
                fit_reg = True,
                
                height = 13,              
                aspect = 1.4      #  dimension of x-axis (sets width)
               )

#plt.margins(x = 2.5, y = 1)
#plt.margins(x = 15, y = 5)


x_col_min = SerieA_table_df['GF'].min()
x_col_max = SerieA_table_df['GF'].max()

y_col_min = SerieA_table_df['GA'].min()
y_col_max = SerieA_table_df['GA'].max()

plt.xlim((x_col_min - 1), (x_col_max + 1))
plt.ylim((y_col_min - 1), (y_col_max + 1))


plt.title('Serie A (20/21) - first seven matches - Goals FOR vs AGAINST',
          fontsize = 40
         )
plt.xlabel('Goals FOR',
           fontsize = 30
          )
plt.ylabel('Goals AGAINST',
           fontsize = 30
          )


point_labels = [plt.text(SerieA_table_df.GF[i],
                         SerieA_table_df.GA[i],
                         
                         SerieA_table_df.Squad[i],
                         fontsize = 15
                         
                        ) for i in range(len(SerieA_table_df.GF))
               ]


adjust_text(point_labels,
            arrowprops = dict(arrowstyle = '->',
                              color = 'red')
           )
```

## Scatterplot - EXPECTED Goals FOR vs EXPECTED Goals AGAINST

```python
ax = sns.lmplot('xG',                       # horizontal axis
                'xGA',                       # vertical axis
                
                data = SerieA_table_df,     # data source
                
                #fit_reg = False,            # do NOT fix a regression line
                fit_reg = True,
                
                height = 10,              
                aspect = 1.3      #  dimension of x-axis (sets width)
               )

#ax.margins(x = 1.5, y = 0.1)
plt.margins(x = 2.5, y = 1)


x_col_min = SerieA_table_df['xG'].min()
x_col_max = SerieA_table_df['xG'].max()

y_col_min = SerieA_table_df['xGA'].min()
y_col_max = SerieA_table_df['xGA'].max()

plt.xlim((x_col_min - 1), (x_col_max + 1))
plt.ylim((y_col_min - 1), (y_col_max + 1))


plt.title('Serie A (20/21) - first seven matches - xG vs xGA',
          fontsize = 40
         )
plt.xlabel('EXPECTED Goals FOR',
           fontsize = 25
          )
plt.ylabel('EXPECTED Goals AGAINST',
           fontsize = 15
          )


point_labels = [plt.text(SerieA_table_df.xG[i],
                         SerieA_table_df.xGA[i],
                         SerieA_table_df.Squad[i]
                        ) for i in range(len(SerieA_table_df.xG))]

adjust_text(point_labels)
```
