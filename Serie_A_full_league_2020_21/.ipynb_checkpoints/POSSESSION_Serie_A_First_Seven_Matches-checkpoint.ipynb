## PART (1) — LOAD into DataFrame(s)

```python
Possession_df = pd.read_csv("Squad_Possession.csv")

Possession_df
```

### Columns list:

```python
columns_list = list(
                    Possession_df.columns
                   )
columns_list
```

## PART (2) — identify RANKINGS per column

```python
Possession_Ranking_df = Possession_df.rank(ascending = 0, # 'False'
                                           method = 'min')

Possession_Ranking_df = Possession_Ranking_df.astype(int)


Possession_Ranking_df['Squad'] = list(map(str.upper,
                                          Possession_df['Squad']
                                         ))
Possession_Ranking_df
```

```python
Possession_Ranking_df.set_index('Squad',
                                inplace = True)
Possession_Ranking_df
```

```python
Possession_Ranks_Transposed = Possession_Ranking_df.transpose()

Possession_Ranks_Transposed
```

```python
Possession_Ranks_Transposed.reset_index(inplace = True)

Possession_Ranks_Transposed.rename(columns = {'index':'SQUAD'
                                             },
                                   # order = {
#                                              'OLD' : 'NEW'
#                                            }
                                   inplace = True
                                  )
Possession_Ranks_Transposed
```

```python
Possession_Rankings = Possession_Ranks_Transposed.sort_values('INTER')
Possession_Rankings
```

```python
Possession_Rankings.reset_index(inplace = True,
                                drop = True)
Possession_Rankings
```

```python
Inter_Rankings_df = (
                     Possession_Rankings[['SQUAD',
                                          'INTER']]
                     .rename(
                             columns = {
                                        'SQUAD' : 'Possession_Variable'
                                       }
# renamed column input order ~~~~~~~~> {'OLD' : 'NEW'}
                            )
                     .set_index(
                                'Possession_Variable'
                               )
                    )
Inter_Rankings_df
```

```python
sns.set_context('talk')

Inter_Rankings_df.plot.hist()
```

```python
Inter_Ranking_Value_Counts = (
                              Inter_Rankings_df[
                                                'INTER'
                                               ]
                              .value_counts()
                              
                              .reset_index(
                                           name = 'Distinct_Variable_Count'
                                          )
                              .rename(
                                      columns = {
                                                 'index':'Possession_Stat_Rank'
                                                }
                                     )
                              .sort_values(
                                           'Possession_Stat_Rank'
                                          )
                             )
Inter_Ranking_Value_Counts
```

```python
TOUCHES_by_PITCH_THIRD_df = (Possession_df[['Squad',
                                            'Touches_Def3rd',
                                            'Touches_Mid3rd',
                                            'Touches_Att3rd'
                                           ]]
                             .sort_values('Touches_Att3rd',
                                          ascending = False
                                         )
                             .set_index('Squad')
                            )
TOUCHES_by_PITCH_THIRD_df
```

## PART (3) — actually PLOTTING the data

```python
sns.set_context(
                'poster',
                font_scale = 2.4
               )

fig, ax = plt.subplots(
                       figsize = (
                                  120,  # horizontal WIDTH
                                  60    # vertical HEIGHT
                                 )
                      )

#### set the positions and width for the bars

pos = list(range(
                 len(
                     TOUCHES_by_PITCH_THIRD_df['Touches_Def3rd']
                    )
                ))
width = 0.25


#### create the bars, in position pos, for Y-axis Variable #1
plt.bar(
        pos,
        
        TOUCHES_by_PITCH_THIRD_df['Touches_Def3rd'], 

        width,
        #alpha = 0.9,
        color = 'purple'   #'#EE3224'
       ) 

#### create bars, in position pos + width buffer, for Y-axis Variable #2
plt.bar([
         (p + width
         ) for p in pos
        ],
        
        TOUCHES_by_PITCH_THIRD_df['Touches_Mid3rd'],

        width,
        #alpha = 0.9,
        color = 'green'     # 'yellow'   # 'lightblue'   #'#F78F1E'
       ) 

#### create bars, in position pos + width buffer, for Y-axis Variable #3
plt.bar([
         (p + width*2
         ) for p in pos
        ],
        
        TOUCHES_by_PITCH_THIRD_df['Touches_Att3rd'], 

        width,
        alpha = 0.8,
        color = 'orange'   #'#FFC222'
       ) 

#ax.set_title('Test', fontsize = 60)
plt.title(
          'TOUCHES - Team Totals per Third of the Pitch',
          fontsize = 100,
          color = 'lightblue'    # 'yellow'    # 'goldenrod'
         )

#ax.set_ylabel('Test Y label')
#plt.ylabel('TACKLING - Totals per Third of the Pitch',fontsize = 70)
#plt.xlabel('SQUADS', fontsize = 35, color = 'yellow')

ax.set_xticks(
              [
               (p + width
               ) for p in pos
              ]                  #  , rot = 90
             )

ax.set_xticklabels(TOUCHES_by_PITCH_THIRD_df.index)  # ['Squad'])

plt.xlim(
         min(pos) - (width * 1.5),
         max(pos) + (width * 3.5)
        )

plt.ylim([
          0,
          (max(
               TOUCHES_by_PITCH_THIRD_df['Touches_Mid3rd']
              ) * 1.03)
         ])

plt.legend(
           [
            'DEFENSIVE third',
            'MIDDLE third',
            'ATTACKING third'
           ],
           loc = 'upper right'
          )
plt.grid()

plt.show()
```

```python
sns.set_context(
                'notebook',
                font_scale = 1.6
               )

ax = sns.lmplot(x = 'Touches',                       # horizontal axis
                y = 'Touches_Live',                   # vertical axis
                
                data = Possession_df,     # data source
                
                fit_reg = False,            # do NOT fix a regression line
                #fit_reg = True,
                
                height = 15,              
                aspect = 1.2      #  dimension of x-axis (sets width)
               )

x_col_min = Possession_df['Touches'].min()
x_col_max = Possession_df['Touches'].max()

y_col_min = Possession_df['Touches_Live'].min()
y_col_max = Possession_df['Touches_Live'].max()

plt.xlim(
         (x_col_min * 0.98),
         (x_col_max * 1.005)
        )
plt.ylim(
         (y_col_min * 0.97),
         (y_col_max * 1.03)
        )


plt.title('Serie A (20/21) - first seven matches - TOUCHES - Total vs Live Ball',
          fontsize = 25,
          color = 'goldenrod'
         )

plt.xlabel('Touches - TOTAL',
           fontsize = 25,
           color = 'yellow'
          )
plt.ylabel('Touches - LIVE BALL',
           fontsize = 25,
           color = 'yellow'
          )


point_labels = [plt.text(
                         Possession_df.Touches[i],
                         Possession_df.Touches_Live[i],
                         
                         Possession_df.Squad[i],
                         fontsize = 20
                         
                        ) for i in range(len(Possession_df.Squad))
               ]

adjust_text(point_labels,
            arrowprops = dict(arrowstyle = '->',
                              color = 'red')
           )
```

```python
sns.set_context(
                'notebook',
                font_scale = 1.6
               )

ax = sns.lmplot(x = 'Touches_Att3rd',                       # horizontal axis
                y = 'Touches_Def3rd',                   # vertical axis
                
                data = Possession_df,     # data source
                
                #fit_reg = False,            # do NOT fix a regression line
                fit_reg = True,
                
                height = 15,              
                aspect = 1.4      #  dimension of x-axis (sets width)
               )

x_col_min = Possession_df['Touches_Att3rd'].min()
x_col_max = Possession_df['Touches_Att3rd'].max()

y_col_min = Possession_df['Touches_Def3rd'].min()
y_col_max = Possession_df['Touches_Def3rd'].max()

plt.xlim(
         (x_col_min * 0.98),
         (x_col_max * 1.01)
        )
plt.ylim(
         (y_col_min * 0.97),
         (y_col_max * 1.015)
        )


plt.title('Serie A (20/21) - first seven matches - TOUCHES - Attacking vs Defensive Third',
          fontsize = 25,
          color = 'goldenrod'
         )

plt.xlabel('Touches - ATTACKING third of the pitch',
           fontsize = 25,
           color = 'yellow'
          )
plt.ylabel('Touches - DEFENSIVE third of the pitch',
           fontsize = 25,
           color = 'yellow'
          )


point_labels = [plt.text(
                         Possession_df.Touches_Att3rd[i],
                         Possession_df.Touches_Def3rd[i],
                         
                         Possession_df.Squad[i],
                         fontsize = 20
                         
                        ) for i in range(len(Possession_df.Squad))
               ]

adjust_text(point_labels,
            arrowprops = dict(arrowstyle = '->',
                              color = 'red')
           )
```
