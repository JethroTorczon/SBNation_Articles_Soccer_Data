## PART (2) — identify RANKINGS per column

```python
Squad_Shooting_Ranked_df = Squad_Shooting_df.rank(ascending = 0,
                                           #      '0' = 'False'
                                                  method = 'min')

Squad_Shooting_Ranked_df = Squad_Shooting_Ranked_df.astype(int)


Squad_Shooting_Ranked_df['Squad'] = list(map(str.upper,
                                             Squad_Shooting_df['Squad']
                                            )
                                        )
Squad_Shooting_Ranked_df
```

```python
Squad_Shooting_Ranked_df.set_index('Squad',
                                   inplace = True)
Squad_Shooting_Ranked_df
```

```python
Squad_Shooting_Ranks_Transposed = Squad_Shooting_Ranked_df.transpose()

Squad_Shooting_Ranks_Transposed
```

```python
Squad_Shooting_Ranks_Transposed.reset_index(inplace = True)

Squad_Shooting_Ranks_Transposed.rename(columns = {'index':'SQUAD'
                                                 },
                                     # order = {
#                                                'OLD' : 'NEW'
#                                              }
                                       inplace = True)
Squad_Shooting_Ranks_Transposed
```

```python
Squad_Shooting_Rankings = Squad_Shooting_Ranks_Transposed.sort_values('INTER')
Squad_Shooting_Rankings
```

```python
Squad_Shooting_Rankings.reset_index(inplace = True,
                                    drop = True)
Squad_Shooting_Rankings
```

```python
Inter_Rankings_df = (
                     Squad_Shooting_Rankings[['SQUAD',
                                              'INTER']]
                     .rename(
                             columns = {
                                        'SQUAD' : 'Shooting_Variable'
                                       }
# renamed column input order ~~~~~~~~> {'OLD' : 'NEW'}
                            )
                     .set_index(
                                'Shooting_Variable'
                               )
                    )
Inter_Rankings_df
```

```python
sns.set_context('talk')

Inter_Rankings_df.plot.hist()
```

```python
Inter_Ranking_Value_Counts = (
                              Inter_Rankings_df[
                                                'INTER'
                                               ]
                              .value_counts()
                              
                              .reset_index(
                                           name = 'Distinct_Variable_Count'
                                          )
                              .rename(
                                      columns = {
                                                 'index':'Shooting_Stat_Rank'
                                                }
                                     )
                              .sort_values(
                                           'Shooting_Stat_Rank'
                                          )
                             )
Inter_Ranking_Value_Counts
```
