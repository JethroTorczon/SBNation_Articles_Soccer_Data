## PART (2) — identify RANKINGS per column

```python
Squad_Passing_Ranking_df = Squad_Passing_df.rank(ascending = 0,
                                         #      '0' = 'False'
                                                 method = 'min'
                                                )

Squad_Passing_Ranking_df = Squad_Passing_Ranking_df.astype(int)


Squad_Passing_Ranking_df['Squad'] = list(map(str.upper,
                                             Squad_Passing_df['Squad']
                                            )
                                        )
Squad_Passing_Ranking_df
```

```python
Squad_Passing_Ranking_df.set_index('Squad',
                                   inplace = True)
Squad_Passing_Ranking_df
```

```python
Squad_Passing_Ranks_Transposed = Squad_Passing_Ranking_df.transpose()

Squad_Passing_Ranks_Transposed
```

```python
Squad_Passing_Ranks_Transposed.reset_index(inplace = True)

Squad_Passing_Ranks_Transposed.rename(columns = {'index':'SQUAD'
                                                },
                                     # order = {
#                                                'OLD' : 'NEW'
#                                              }
                                      inplace = True
                                     )
Squad_Passing_Ranks_Transposed
```

```python
Squad_Passing_Rankings = Squad_Passing_Ranks_Transposed.sort_values('INTER')

Squad_Passing_Rankings
```

```python
Squad_Passing_Rankings.reset_index(inplace = True,
                                   drop = True)
Squad_Passing_Rankings
```

```python
Inter_Rankings_df = (
                     Squad_Passing_Rankings[['SQUAD',
                                             'INTER']]
                     .rename(
                             columns = {
                                        'SQUAD' : 'Passing_Variable'
                                       }
# renamed column input order ~~~~~~~~> {'OLD' : 'NEW'}
                            )
                     .set_index(
                                'Passing_Variable'
                               )
                    )
Inter_Rankings_df
```

```python
sns.set_context('talk')

Inter_Rankings_df.plot.hist()
```

```python
Inter_Ranking_Value_Counts = (
                              Inter_Rankings_df[
                                                'INTER'
                                               ]
                              .value_counts()
                              
                              .reset_index(
                                           name = 'Distinct_Variable_Count'
                                          )
                              .rename(
                                      columns = {
                                                 'index':'PASSING_Stat_Rank'
                                                }
                                     )
                              .sort_values(
                                           'PASSING_Stat_Rank'
                                          )
                             )
Inter_Ranking_Value_Counts
```

```python
sns.set_context(
                'notebook',
                font_scale = 1.8
               )

ax = sns.lmplot(x = 'Att',                       # horizontal axis
                y = 'Cmp',                       # vertical axis
                
                data = Squad_Passing_df,     # data source
                
                #fit_reg = False,            # do NOT fix a regression line
                fit_reg = True,
                
                height = 18,              
                aspect = 1.2      #  dimension of x-axis (sets width)
               )

#plt.margins(x = 2.5, y = 1)
#plt.margins(x = 15, y = 5)


x_col_min = Squad_Passing_df['Att'].min()
x_col_max = Squad_Passing_df['Att'].max()

y_col_min = Squad_Passing_df['Cmp'].min()
y_col_max = Squad_Passing_df['Cmp'].max()

plt.xlim((x_col_min * 0.98),
         (x_col_max * 1.025)
        )
plt.ylim((y_col_min * 0.95),
         (y_col_max * 1.025)
        )


plt.title('Serie A (20/21) - first seven matches - Passing ATTEMPTS vs COMPLETIONS',
          fontsize = 25,
          color = 'goldenrod'
         )

plt.xlabel('Passing ATTEMPTS',
           fontsize = 30,
           color = 'yellow'
          )
plt.ylabel('Passing COMPLETIONS',
           fontsize = 30,
           color = 'yellow'
          )


point_labels = [plt.text(Squad_Passing_df.Att[i],
                         Squad_Passing_df.Cmp[i],
                         
                         Squad_Passing_df.Squad[i],
                         fontsize = 25
                         
                        ) for i in range(len(Squad_Passing_df.Squad))
               ]


adjust_text(point_labels,
            arrowprops = dict(arrowstyle = '->',
                              color = 'red')
           )
```
