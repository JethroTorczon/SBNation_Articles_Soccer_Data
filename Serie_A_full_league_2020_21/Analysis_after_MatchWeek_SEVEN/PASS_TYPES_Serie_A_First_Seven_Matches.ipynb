## PART (2) — identify RANKINGS per column

```python
Squad_Pass_Types_Ranking_df = Squad_Pass_Types_df.rank(ascending = 0,
                                                 #     '0' = 'False'
                                                       method = 'min')

Squad_Pass_Types_Ranking_df = Squad_Pass_Types_Ranking_df.astype(int)


Squad_Pass_Types_Ranking_df['Squad'] = list(map(str.upper,
                                             Squad_Pass_Types_df['Squad']
                                               )
                                           )
Squad_Pass_Types_Ranking_df
```

```python
Squad_Pass_Types_Ranking_df.set_index('Squad',
                                      inplace = True)
Squad_Pass_Types_Ranking_df
```

```python
Squad_Pass_Types_Ranks_Transposed = Squad_Pass_Types_Ranking_df.transpose()

Squad_Pass_Types_Ranks_Transposed
```

```python
Squad_Pass_Types_Ranks_Transposed.reset_index(inplace = True)

Squad_Pass_Types_Ranks_Transposed.rename(columns = {'index':'SQUAD'
                                                   },
                                       #  order = {
#                                                  'OLD' : 'NEW'
#                                                 }
                                         inplace = True
                                        )
Squad_Pass_Types_Ranks_Transposed
```

```python
Squad_Pass_Types_Rankings = Squad_Pass_Types_Ranks_Transposed.sort_values('INTER')

Squad_Pass_Types_Rankings
```

```python
Squad_Pass_Types_Rankings.reset_index(inplace = True,
                                      drop = True)
Squad_Pass_Types_Rankings
```

```python
Inter_Rankings_df = (
                     Squad_Pass_Types_Rankings[['SQUAD',
                                                'INTER']]
                     .rename(
                             columns = {
                                        'SQUAD' : 'Pass_Types_Variable'
                                       }
# renamed column input order ~~~~~~~~> {'OLD' : 'NEW'}
                            )
                     .set_index(
                                'Pass_Types_Variable'
                               )
                    )
Inter_Rankings_df
```

```python
sns.set_context('talk')

Inter_Rankings_df.plot.hist()
```

```python
Inter_Ranking_Value_Counts = (
                              Inter_Rankings_df[
                                                'INTER'
                                               ]
                              .value_counts()
                              
                              .reset_index(
                                           name = 'Distinct_Variable_Count'
                                          )
                              .rename(
                                      columns = {
                                                 'index':'PASS_TYPE_Stat_Rank'
                                                }
                                     )
                              .sort_values(
                                           'PASS_TYPE_Stat_Rank'
                                          )
                             )
Inter_Ranking_Value_Counts
```

```python
sns.set_context(
                'notebook',
                font_scale = 1.8
               )

ax = sns.lmplot(x = 'Press',                       # horizontal axis
                y = 'Cmp',                       # vertical axis
                
                data = Squad_Pass_Types_df,     # data source
                
                #fit_reg = False,            # do NOT fix a regression line
                fit_reg = True,
                
                height = 18,              
                aspect = 1.2      #  dimension of x-axis (sets width)
               )

x_col_min = Squad_Pass_Types_df['Press'].min()
x_col_max = Squad_Pass_Types_df['Press'].max()

y_col_min = Squad_Pass_Types_df['Cmp'].min()
y_col_max = Squad_Pass_Types_df['Cmp'].max()

plt.xlim((x_col_min * 0.98),
         (x_col_max * 1.025)
        )
plt.ylim((y_col_min * 0.95),
         (y_col_max * 1.025)
        )


plt.title('Serie A (20/21) - first seven matches - Complete Passes - Under PRESSURE vs TOTAL',
          fontsize = 25,
          color = 'goldenrod'
         )

plt.xlabel('Passing COMPLETIONS - under PRESSURE',
           fontsize = 30,
           color = 'yellow'
          )
plt.ylabel('Passing COMPLETIONS - TOTAL',
           fontsize = 30,
           color = 'yellow'
          )


point_labels = [plt.text(Squad_Pass_Types_df.Press[i],
                         Squad_Pass_Types_df.Cmp[i],
                         
                         Squad_Pass_Types_df.Squad[i],
                         fontsize = 25
                         
                        ) for i in range(len(Squad_Pass_Types_df.Squad))
               ]


adjust_text(point_labels,
            arrowprops = dict(arrowstyle = '->',
                              color = 'red')
           )
```
