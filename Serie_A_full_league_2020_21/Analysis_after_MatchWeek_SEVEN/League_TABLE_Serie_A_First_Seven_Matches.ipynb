## PART (1) — LOAD into DataFrame(s)

```python
SerieA_table = pd.read_csv("Serie_A_table_after_7_matches.csv")

SerieA_table_df = SerieA_table.loc[:,
                                   'Rk':'xGDiff/90']
SerieA_table_df
```

### Columns list:

```python
columns_list = list(
                    SerieA_table_df.columns
                   )
columns_list
```

## PART (2) — identify RANKINGS per column

```python
SerieA_table_Ranking_df = SerieA_table_df.rank(ascending = 0,
                                         #      '0' = 'False'
                                               method = 'min'
                                              )

SerieA_table_Ranking_df = SerieA_table_Ranking_df.astype(int)


SerieA_table_Ranking_df['Squad'] = list(map(str.upper,
                                            SerieA_table_df['Squad']
                                           )
                                       )
SerieA_table_Ranking_df
```

```python
SerieA_table_Ranking_df.set_index('Squad',
                                  inplace = True)
SerieA_table_Ranking_df
```

```python
SerieA_table_Ranks_Transposed = SerieA_table_Ranking_df.transpose()

SerieA_table_Ranks_Transposed
```

```python
SerieA_table_Ranks_Transposed.reset_index(inplace = True
                                         )
SerieA_table_Ranks_Transposed.rename(columns = {'index':'SQUAD'
                                               },
                                     # order = {
#                                                'OLD' : 'NEW'
#                                              }
                                     inplace = True
                                    )
SerieA_table_Ranks_Transposed
```

```python
SerieA_table_Rankings = SerieA_table_Ranks_Transposed.sort_values('INTER')
SerieA_table_Rankings
```

```python
SerieA_table_Rankings.reset_index(inplace = True,
                                  drop = True)
SerieA_table_Rankings
```

```python
Inter_Rankings_df = (
                     SerieA_table_Rankings[['SQUAD',
                                            'INTER']]
                     .rename(
                             columns = {
                                        'SQUAD' : 'Defensive_Variable'
                                       }
# renamed column input order ~~~~~~~~> {'OLD' : 'NEW'}
                            )
                     .set_index(
                                'Defensive_Variable'
                               )
                    )
Inter_Rankings_df
```

```python
sns.set_context('talk')

Inter_Rankings_df.plot.hist()
```

```python
Inter_Ranking_Value_Counts = (
                              Inter_Rankings_df[
                                                'INTER'
                                               ]
                              .value_counts()
                              
                              .reset_index(
                                           name = 'Distinct_Variable_Count'
                                          )
                              .rename(
                                      columns = {
                                                 'index':'Standard_Stat_Rank'
                                                }
                                     )
                              .sort_values(
                                           'Standard_Stat_Rank'
                                          )
                             )
Inter_Ranking_Value_Counts
```

```python
ax = sns.lmplot(x = 'GF',                       # horizontal axis
                y = 'GA',                       # vertical axis
                
                data = SerieA_table_df,     # data source
                
                #fit_reg = False,            # do NOT fix a regression line
                fit_reg = True,
                
                height = 13,              
                aspect = 1.4      #  dimension of x-axis (sets width)
               )

#plt.margins(x = 2.5, y = 1)
#plt.margins(x = 15, y = 5)


x_col_min = SerieA_table_df['GF'].min()
x_col_max = SerieA_table_df['GF'].max()

y_col_min = SerieA_table_df['GA'].min()
y_col_max = SerieA_table_df['GA'].max()

plt.xlim((x_col_min - 1), (x_col_max + 1))
plt.ylim((y_col_min - 1), (y_col_max + 1))


plt.title('Serie A (20/21) - first seven matches - Goals FOR vs AGAINST',
          fontsize = 25,
          color = 'goldenrod'
         )

plt.xlabel('Goals FOR',
           fontsize = 25,
           color = 'yellow'
          )
plt.ylabel('Goals AGAINST',
           fontsize = 25,
           color = 'yellow'
          )


point_labels = [plt.text(SerieA_table_df.GF[i],
                         SerieA_table_df.GA[i],
                         
                         SerieA_table_df.Squad[i],
                         fontsize = 25
                         
                        ) for i in range(len(SerieA_table_df.GF))
               ]


adjust_text(point_labels,
            arrowprops = dict(arrowstyle = '->',
                              color = 'red')
           )
```
