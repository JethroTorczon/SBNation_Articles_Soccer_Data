# COLUMN CLEANING custom FUNCTIONS

```python
def rm_first_word(raw_column_list):
    #
    col_single_split = raw_column_list.str.split( # pat = " ",
                                                 n = 1,
                                                 expand = True
                                                )
    #
    cleaned_list = col_single_split[1]
    #
    
    return cleaned_list
```

```python
def rm_backslash_thru_end(raw_column_list):
    #
    col_single_split = raw_column_list.str.split(pat = "\\",
                                                 n = 1,
                                                 expand = True
                                                )
    #
    cleaned_list = col_single_split[0]
    #
    
    return cleaned_list
```

# Player DF (1) - "STANDARD" stats

```python
file_name_1 = "STANDARD_stats.csv"
```

```python
df_1 = pd.read_csv(
                   folder_path +
                   file_name_1
                  )
```

```python
# Standard_df

df_1
```

# CLEANING "STRING" COLUMNS

```python
nation_column = df_1["Nation"]
```

## Method (1)  -  str.PARTITION

```python
# nation_column = df_1["Nation"]
# nation_column
```

```python
# nation_col_partition = nation_column.str.partition()
# nation_col_partition
```

```python
# # df_1["Nation"] = nation_col_partition[:,2]

# Nation_clean = nation_col_partition[2]
# Nation_clean
```

```python
# nation_column = df_1["Nation"]

nation_col_partition = nation_column.str.partition()

Nation_clean = nation_col_partition[2]

Nation_clean
```

```python
df_1["Nation"] = Nation_clean

df_1.head(3)
```

## Method (2)  -  str.SPLIT

```python
league_column = df_1["Comp"]

# league_column
```

```python
# league_col_split = league_column.str.split()

league_col_SINGLE_split = league_column.str.split(n = 1,
                                                  expand = True)
# league_col_SINGLE_split
```

```python
League_clean = league_col_SINGLE_split[1]

# League_clean
```

```python
df_1["Comp"] = League_clean

df_1.head()
```

```python
Player_column = df_1["Player"]

Player_col_SINGLE_split = Player_column.str.split(pat = '\\',
                                                  n = 1,
                                                  expand = True)
Player_clean = Player_col_SINGLE_split[0]

df_1["Player"] = Player_clean

# df_1.head()
```

```python
df_1
```

```python
df_1.to_csv(clean_data_folder +
            file_name_1,
            
            index = False
           )
```
